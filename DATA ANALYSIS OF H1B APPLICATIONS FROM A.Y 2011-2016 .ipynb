```python
#DATA ANALYSIS OF H1B PETITIONS from 2011-2016
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('F:/Data_Science/H1B.csv')
import seaborn as sns
```

```python
df['CASE_STATUS']=df['CASE_STATUS'].apply(lambda x: x.upper())
df['EMPLOYER_NAME']=df['EMPLOYER_NAME'].apply(lambda x: x.upper())
df['SOC_NAME']=df['SOC_NAME'].apply(lambda x: x.upper())
df['JOB_TITLE']=df['JOB_TITLE'].apply(lambda x: x.upper())
df['FULL_TIME_POSITION']=df['FULL_TIME_POSITION'].apply(lambda x: x.upper())
df['WORKSITE']=df['WORKSITE'].apply(lambda x: x.upper())
```

```python
df['EMPLOYER_NAME'].value_counts().sort_values(ascending=False).head(15)
```

```python
plt.figure(figsize=(15,9))
df['EMPLOYER_NAME'].value_counts().sort_values(ascending=False).head(15).plot(kind=('bar'), title='TOP 15 HIRING COMPANIES',alpha=0.6,rot=90,color='r')
plt.xlabel('EMPLOYER NAME')
plt.ylabel('NO OF EMPLOYEES HIRED')
plt.show()
```

```python
plt.figure(figsize=(15,8))
explode=[0,0.2,0,0.2]
plt.axis('equal')
x=df['CASE_STATUS'].value_counts().head(4)
df['CASE_STATUS'].value_counts().head(4).plot(kind='pie',explode=explode,radius=1,autopct=lambda p:'{:.1f}%\n({:.0f})'.format(p,(p/100)*x.sum()), title='H1B VISA STATS')
```

```python
plt.figure(figsize=(15,7))
df.groupby(['YEAR'])['YEAR'].count().sort_values().plot(kind='bar', title='NO OF PETITIONS FILED',color='orange',alpha=0.8,rot=90)
plt.xlabel('YEAR')
plt.ylabel('NO OF PETITIONS FILED')
plt.show()
```

```python
plt.figure(figsize=(12,6))
df[df['CASE_STATUS']=='DENIED']['YEAR'].value_counts().sort_index().plot(kind='bar',color='r',alpha=0.6)
plt.ylabel('No of Denied Petitions',color='blue',fontsize='large')
plt.xlabel('Year',color='blue',fontsize='large')
plt.xticks(color='blue')
plt.legend(shadow=True)
plt.title('DENIED PETITIONS',color='red',fontsize='large')
plt.show()
```

```python
plt.figure(figsize=(10,5))

count=df['YEAR'].value_counts().sort_index()
(((df[df['CASE_STATUS']=='DENIED']['YEAR'].value_counts().sort_index())/count)*100).plot(kind='bar',color='r',alpha=0.6)

plt.xticks(color='blue')
plt.ylabel('Rate of Denied Petitions(%)',color='blue',fontsize='large')
plt.xlabel('Year',color='blue',fontsize='large')
plt.legend(shadow=True)
plt.title('RATE OF DENIED PETITIONS(%)',color='red',fontsize='large')
```
